For (N, m, m - 1) model, that picking m cards from N cards and display one card, we show that as long as N <= m! + m - 1, there is a memorizable strategy

# M Random Numbers

```python
# function to compute the least number m which is enough for (N, m, m - 1) trick
def compute_m(N):
  i = 1
  factorial = 1
  while factorial + i - 1 < N:
    i += 1
    factorial = factorial * i
  return i
```

```python
# input is an arbitrary number N representing the total number of cards
N = 120
m = compute_m(N)
print(m)
```

```python
# test the algorithm
# generate m random numbers from 1-N

import random

def generate_m_number(N, m):
  s = set()
  for i in range(m):
    r = random.randint(1, N)
    while r in s:
      r = random.randint(1, N)
    s.add(r)
  return s
```

# Permuation


```python
import math
# permutation function using the permuation of cards to represent a number k
def permuate(cards, k):
  print("the card before permuate is " + str(cards) + " and k is " + str(k))
  p = []
  l = len(cards)
  # this k is starting from 0
  index = k + 1
  for i in range(1, l):
    d = index // math.factorial(l - i)
    r = index % math.factorial(l - i)
    if r == 0:
      p.append(cards[d - 1])
      del cards[d - 1]
      p.extend(cards[::-1])
      return p
    else:
      p.append(cards[d])
      del cards[d]
    index = r
  return p
```

```python
# de permutation function, construct a number from permuations
def de_permuate(cards):
  k = 0
  sorted_cards = sorted(cards)
  for i in range(len(cards)):
    card = cards[i]
    index = sorted_cards.index(card)
    sorted_cards.remove(card)
    k = k + index * math.factorial(len(cards) - i - 1);
  return k
```

# Find Index of the hiden cards

```python
# given the displayed cards generate the possible cards list
# and find the index of the hiden in the possible cards list
def find_index(cards, hiden, m, N):
  possible_cards = []
  for i in range(m - 1):
    r = (m - (sum(cards) % m) + i) % m
    card = r
    lower_bound = 0
    if i > 0:
      lower_bound = cards[i - 1]
    while card < cards[i]:
      if (card > lower_bound):
        possible_cards.append(card)
      card = card + m
  # i = m - 1
  r = (m - (sum(cards) % m) + m - 1) % m
  card = r
  while card <= N:
      if card > cards[m - 2]:
        possible_cards.append(card)
      card = card + m
  print("the possible_cards is " + str(possible_cards))
  k = possible_cards.index(hiden)

  return k
```

# Display and Decrypt


```python
def display_and_hide_cards(cards, N):
  # compute the reminder, and hide the reminder + 1th card
  m = len(cards)
  reminder = sum(cards) % m;
  hiden = cards[reminder];
  print("The sum is " + str(sum(cards)) + ", The reminder is " + str(reminder) + ", so hide the " + str(hiden))

  # permutate the remining cards to represent k
  # generate the possible hiden card list
  cards.remove(hiden)

  k = find_index(cards, hiden, m, N);
  print("the index is " + str(k))
  cards = permuate(cards, k)
  print("the permuated cards is " + str(cards))
  return cards, hiden
```

```python
# decrypt the hiden using the displayed cards
def decrypt(cards):
   m = len(cards) + 1
   k = de_permuate(cards);
   print("the de_premuated k is " + str(k))
   r = m - (sum(cards) % m)
   card = r + m * k
   index = m - 1
   sorted_cards = sorted(cards)
   for i in range(m - 1):
    if card < sorted_cards[i]:
      index = i
      break
    card = card + 1
   print("the card is " + str(card) + " and the index is " + str(index))
   return card
```
