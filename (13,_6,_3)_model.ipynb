# Let's talk about the case that m = 6, the upper bound is 13

```python
#U: Display of k(3) cards, (13 choose 3) * 3! = 56.
#V: a set of m(6) cards, (13 choose 6) = 56.
#|U| >= |V|.

import math

n = 13
# 计算组合数
number_u = math.comb(n, 3) * 6
number_v = math.comb(n, 6)

print("number_u: ", number_u, " number_v: ", number_v)
```

```python
from itertools import combinations
from itertools import permutations

cards = [_ for _ in range(1, n + 1)]

# 获取所有有序取出3张牌的排列
pers = list(permutations(cards, 3))
print(len(pers))

per_set = set(pers)

# 获取所有有序取出6张牌的组合
combs = list(combinations(cards, 6))
print(len(combs))

comb_set = set(combs)
# 输出结果
#for combination in combinations:
#    print(combination)
```

# Use HOO method to get a comb from a permutation

```python
import copy
import bisect

def recover_3_to_6(per):
  index_3 = 0
  index_2 = 0
  # 123
  if per[0] < per[1] and per[1] < per[2] and per[0] < per[2]:
    index_3 = 0
    index_2 = 0
  # 132
  if per[0] < per[1] and per[1] > per[2] and per[0] < per[2]:
    index_3 = 1
    index_2 = 0
  # 213
  if per[0] > per[1] and per[1] < per[2] and per[0] < per[2]:
    index_3 = 2
    index_2 = 0
  # 231
  if per[0] < per[1] and per[1] > per[2] and per[0] > per[2]:
    index_3 = 0
    index_2 = 1
  # 312
  if per[0] > per[1] and per[1] < per[2] and per[0] > per[2]:
    index_3 = 1
    index_2 = 1
  # 321
  if per[0] > per[1] and per[1] > per[2] and per[0] < per[2]:
    index_3 = 2
    index_2 = 1
  #print("index_3:", index_3, ", index_2:", index_2)
  temp_list = copy.deepcopy(list(per))
  temp_list.sort()

  # 3 => 4
  potential_4 = []
  for i in range(1, n + 1):
    if i in per:
      continue
    sum_list = sum(temp_list) + i
    reminder = sum_list % 4
    location = bisect.bisect_left(temp_list, i)
    if reminder == location:
      potential_4.append(i)

  if index_3 + 1 > len(potential_4):
    #print("index_3 + 1 > len(potential_4), index_3: ", index_3, "potential_4: ", potential_4)
    #print(temp_list)
    return

  temp_list.append(potential_4[index_3])
  temp_list.sort()
  #print("3 => 4: ", temp_list, "index_3: ", index_3)

  # 4 => 5
  potential_5 = []
  for i in range(1, n + 1):
    if i in temp_list:
      continue
    sum_list = sum(temp_list) + i
    reminder = sum_list % 5
    location = bisect.bisect_left(temp_list, i)
    if reminder == location:
      potential_5.append(i)

  if index_2 + 1 > len(potential_5):
    #print("index_2 + 1 > len(potential_5), index_2: ", index_2, "potential_5: ", potential_5)
    #print(temp_list)
    return
  temp_list.append(potential_5[index_2])
  temp_list.sort()
  #print("4 => 5: ", temp_list)

  # 5 => 6
  for i in range(1, n + 1):
    if i in temp_list:
      continue
    sum_list = sum(temp_list) + i
    reminder = sum_list % 6
    location = bisect.bisect_left(temp_list, i)
    if reminder == location:
      temp_list.append(i)
      break

  temp_list.sort()
  #print("5 => 6: ", temp_list)
  return temp_list
```
